```python
import pandas as pd
import numpy as np
%matplotlib inline
from ggplot import *
import random
import matplotlib.pyplot as plt

rat_cols = ['uid', 'mid', 'r', 'timestamp']
rating_data = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.data', sep = '\t', names=rat_cols)
print rating_data.head()

m_cols = ['mid', 'title', 'rdate','vrdate', 'imdburl']
movie_data = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.item', sep = '|', names=m_cols, usecols=range(5))
print movie_data.head()
```

```python
mergedata = pd.merge(rating_data,movie_data, on='mid')
print mergedata.head()
```

```python
movie_pt=mergedata.pivot_table(index=['uid'], columns =['title'], values=['r'])
movie_pt.columns= movie_pt.columns.droplevel(0)
movie_pt.head()
```

```python
starwarsratings=movie_pt['Star Wars (1977)']
starwarsratings.head()
```

```python
simmovies=movie_pt.corrwith(starwarsratings)
simmovies=simmovies.dropna()
simmovies.head()
```

```python
z=pd.DataFrame(simmovies.sort_values(ascending=False))
#z.rows = z.rows.droplevel(0)
#z=pd.DataFrame()
print z.head()
type(z)
```

```python
movie_gp = mergedata.groupby(['title']).agg({'r': {'r_std': np.std, 'r_mean': np.mean, 'r_count': np.size}})
movie_gp.columns = movie_gp.columns.droplevel(0)
a=movie_gp['r_count']>50
b=movie_gp[a]
print b.head()
type(b)
```

```python
sim_100_mov= b.join(z)
sim_100_mov.columns = ['count', 'mean_ratings', 'std_ratings', 'sim']
#print sim_100_mov.head(200)
#type (sim_100_mov)
sim_100_mov.sort_values('sim', ascending=False)
```
